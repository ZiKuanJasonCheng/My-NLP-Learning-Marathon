## 作業目標：搭建一個bag of words模型

---

```python
import pandas as pd
import nltk

import numpy as np
dataset = pd.read_csv('Restaurant_Reviews.tsv', delimiter='\t', quoting=3)  # quoting=QUOTE_NONE
corpus = dataset['Review'].values
print(len(corpus))
corpus
```

### 從文本中取出所有單字

```python
whole_words = []
for sentence in corpus:
    tokenized_sentence = '使用tokenize 技巧'
    for word in tokenized_sentence:
        whole_words.append(word)
```

```python
from nltk.tokenize import word_tokenize

whole_words = []
for sentence in corpus:
    tokenized_sentence = word_tokenize(sentence)  #'使用tokenize 技巧'
    for word in tokenized_sentence:
        whole_words.append(word)
whole_words
```

### 移除重複單字

```python
whole_words = list(set(whole_words))
print('There are {} words in total.'.format(len(whole_words)))
```

### 建立字典使每一個單字有對應數值

```python
word_index = {}
index_word = {}
n = 0
for word in whole_words:
    """自行填入"""
    index_word[n] = word
    word_index[word] = n
    n += 1
    
```

## 轉換句子為bag of words型式

```python
from sklearn.feature_extraction.text import CountVectorizer

def _get_bag_of_words_vector(sentence, word_index_dic, whole_words):
    sentence = sentence
    '''創建一個vector'''
    #cv = CountVectorizer(max_features=len(whole_words))
    #cv.fit_transform(corpus).toarray()
    vector = np.zeros((len(whole_words)))
    for word in nltk.word_tokenize(sentence):
        if word in whole_words:
            vector[word_index_dic[word]] = 1
    #vector = cv.transform([sentence]).toarray()
    return vector
```
